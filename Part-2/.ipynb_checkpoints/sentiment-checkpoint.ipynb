```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import json
import requests
from pprint import pprint
import tweepy
from datetime import datetime
import numpy as np

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Import API keys
from config import api_key, api_secret_key, access_token, access_token_secret
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(api_key, api_secret_key)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target User
media = ["BBC","CBS", "CNN", "Fox", "nytimes"]

# Tweet Texts
tweet_texts = []
media_source = []
tweet_times = []

# Get all tweets from home feed
for x in range(1, 6):
    for source in media:
        public_tweets = api.user_timeline(source, page=x)
        
        # Loop through all tweets
        for tweet in public_tweets:
            raw_time = tweet["created_at"]
            text = tweet["text"]
            # Store Tweet in Array
            tweet_texts.append(text)
            media_source.append(source)
            tweet_times.append(raw_time)
```

```python
# Create dataframe
dict = {'media': media_source, 'text': tweet_texts, 'time': tweet_times}  
tweets_df = pd.DataFrame(dict)
len(tweets_df)
```

```python
# Run analysis
sentiment = []
neg = []
pos = []
neu = []
comp = []
for text in tweets_df["text"]:
    sentiment.append(analyzer.polarity_scores(text))
for i in range(len(sentiment)):
    neg.append(sentiment[i]["neg"])
    neu.append(sentiment[i]["neu"])
    pos.append(sentiment[i]["pos"])
    comp.append(sentiment[i]["compound"])
tweets_df["neg"] = neg
tweets_df["neu"] = neu
tweets_df["pos"] = pos
tweets_df["compound"] = comp
tweets_df[350:370]
tweets_df = tweets_df.set_index("media")
```

```python
# Sort by time
tweets_df = tweets_df.sort_values(["time"])
tweets_df

# Export dataframe 
tweets_df.to_csv("sentiment_data.csv")
```

```python
# Group dataframe by media
groups = tweets_df.groupby("media")
group_compound = groups["compound"].mean()
group_compound
```

```python
# Setup the scatter plot axises
x_axis = range(100)
tick_locations = [0, 20, 40, 60, 80, 100]
plt.title("Sentiment analysis of media tweets (11/05/2019)")
plt.xlabel("Tweets ago")
plt.ylabel("Tweet Polarity")

plt.xticks(tick_locations, [100,80, 60,40,20,0]) 
```

```python
# Scatter plot with labels
plt.scatter(x_axis, tweets_df.loc["BBC","compound"], color="c", edgecolor = "black", alpha = 0.75,label="BBC")

plt.scatter(x_axis, tweets_df.loc["CBS","compound"], color="g", edgecolor = "black", alpha = 0.75,label="CBS")

plt.scatter(x_axis, tweets_df.loc["CNN","compound"], color="r", edgecolor = "black", alpha = 0.75,label="CNN")

plt.scatter(x_axis, tweets_df.loc["Fox","compound"], color="b", edgecolor = "black", alpha = 0.75,label="Fox")

plt.scatter(x_axis, tweets_df.loc["nytimes","compound"], color="y", edgecolor = "black", alpha = 0.75,label="New York Times")

# Place a legend on the chart
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.grid()

# Save figure
plt.savefig("Images/sentiment_scatter.png")
```
